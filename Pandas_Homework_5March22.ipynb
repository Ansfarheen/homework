## For titanic datset :

### 6. how many person we have whose name starts with "s"

```python
len(df[df['Name'].apply(lambda x : x.startswith('s') or x.startswith('S'))])
```

### 9 . how many person died whose age was less then 40

### 10 . from a  cabin column seperate text and numeric value . 

```python
df.dropna(inplace=True)
df['Cabin_let']=[x[0] for x in df['Cabin']]
df['Cabin_num']=[x[1:] for x in df['Cabin']]
df
```
