# Create a database called as cardataset.

```python
import mysql.connector as connection
import logging

formatter= logging.Formatter('%(asctime)s:%(levelname)s: %(name)s : %(message)s')

logger = logging.getLogger(__name__)
logger.setLevel(logging.INFO)
file_handler=logging.FileHandler('mysql.log')
file_handler.setFormatter(formatter)
logger.addHandler(file_handler)
stream_handler=logging.StreamHandler()
stream_handler.setFormatter(formatter)
stream_handler.setLevel(logging.INFO)
logger.addHandler(stream_handler)

try:
    mydb= connection.connect(host='localhost',user='root',passwd='Farheen@94',use_pure=True)
    query= "create database cardataset"
    cur= mydb.cursor()
    cur.execute(query)
    
except Exception as e:
    logger.error(e)
    
else:
    logger.info('cardataset database has been created.')
if mydb.is_connected():
    mydb.close()
```

# Create a table called as car with a column name given in dataset description.

```python
try:
    mydb= connection.connect(host='localhost',user='root',passwd='Farheen@94',database='cardataset',use_pure=True)
    query= "create table car(buying varchar(10),maint varchar(10),doors varchar(10),persons varchar(10),lug_boot varchar(6),safety varchar(5),class varchar(7))"

    cur= mydb.cursor()
    cur.execute(query)
    
except Exception as e:
    logger.error(e)
    
else:
    logger.info('car table has been created.')

if mydb.is_connected():
    mydb.close()
```

# Dump all the data into car table.

```python
import csv
try:
    mydb= connection.connect(host='localhost',user='root',passwd='Farheen@94',database='cardataset',use_pure=True)
    cur= mydb.cursor()
    
    with open('car.data','r') as fd:
        car_data=csv.reader(fd,delimiter='\n')
        for i in car_data:
            k= str(i[0].split(','))
            cur.execute(f'insert into car values({k[1:-1]})')
    mydb.commit() 
    
except Exception as e:
    logger.error(e)
    
else:
    logger.info(f'Data has been dumped into car table.')

if mydb.is_connected():
    mydb.close()
```
